```python
## 100이하의 정수만 입력. while문 이용해서 1부터 입력받은 수 까지 합


num = int(input("100이하의 정수를 입력"))

i = 1
sum = 0

while i <= num:
  sum = sum + i
  i = i + 1

print(sum)
  
```

```python
## 100이하의 정수를 입력받아서 입력받은 정수부터 100까지의 합을 출력.

num = int(input("100 이하의 정수를 입력"))

i = num
sum = 0

while i <= 100:
  sum = sum + i
  i = i + 1

print(sum)

```

```python
## 0이 입력될때 까지 정수를 입력받아 3의 배수와 5의 배수를 제외한 수들의 개수를 출력

sum = 0


while True:
  num = int(input("숫자를 입력하시오"))
  if num == 0:
    break
  else:
    if num%3 != 0 and num%5 != 0:
        sum = sum +1

print(sum)

```

```python
## 삼각형의 밑변의 길이와 높이를 입력받아 넓이를 출력, continue 물어봐서 y,Y 면 다시 입력, 다른 문자면 종료


while True:
  base = input("밑변 길이")
  base = int(base)

  height = input("높이")
  height = int(height)

  triangleArea = (base * height) / 2

  print(triangleArea)

  question = input("continue?")
  if question == "Y":
    print()
  elif question == "y":
    print()
  else:
    break 
```
