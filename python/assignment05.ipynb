```python
##자연수 n 을 입력받고, 1부터 홀수를 차례대로 더해나가면서 합이 n 이상이면 // 그 때까지 더해진 홀수의 개수와 그 합을 출력

number = int(input("자연수"))

i = 1
li = []

while sum(li)< number:
  li.append(i)
  i += 2
print(len(li), sum(li))
```

```python
## 2부터 9까지의 수 중 2개를 입력받아 입력받은 수 사이의 구구단을 나오게 출력
## 반드시 먼저 입력된 수의 구구단 부터 구구단 사이의 공백은 3칸이다.

x = int(input("첫번째 숫자"))
y = int(input("두번째 숫자"))

if x <= y:
  for i in range(1, 10):
    for j in range(x, y+1, 1):
      print(j, "*", i, "=", i*j, end='  ')
    print()
else:
  for i in range(1,10):
    for j in range(x, y-1, -1):
      print(j, "*", i, "=", i*j, end='  ')
    print()
```

```python
## 행과 열의 수를 입력받는다

col = int(input("열"))
row = int(input("행"))


for i in range(1, row+1):
  for j in range(1, col+1):
    print(i*j, end=' ')
  print()
```
