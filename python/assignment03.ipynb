```python
#정수입력 0 이면 z, 양p, 음m 출력

num = input()
num = int(num)

if num > 0:
  print("plus")
elif num == 0:
  print("zero")
else:
  print("minus")
```

```python
word = input()

if word == "A":
  print("Excellent")
elif word == "B":
  print("Good")
elif word == "C":
  print("Usually")
elif word == "D":
  print("Effort")
elif word == "E":
  print("Failure")
else:
  print("Error")
```

```python

# 1,3,5,7,8,10,12 = 31
# 4,6,9,11 = 30
# 2 = 28

month = input()
month = int(month)

if month == 1 or 3 or 5 or 7 or 8 or 10 or 12:
  print("31")
elif month == 4 or 6 or 9 or 11:
  print("30")
else:
  print("28")



```

```python
sex = input("성별을 입력하시오:")
age = input("나이를 입력하시오:")
age = int(age)

if sex == "M" and age >= 18:
  print("MAN")
elif sex == "M" and age < 18:
  print("BOY")
elif sex == "F" and age >= 18:
  print("WOMAN")
elif sex == "F" and age < 18:
  print("GIRL")
else:
  print("NOTHING")

```

```python
#leap 윤년, common 평년
# 400으로 나누어떨어지면 윤년 or 4로 나누어 떨어지고 100으로 나누어떨어지지 않으면 윤년

year = input("년도를 입력하시오:")
year = int(year)

if year % 400 == 0:
  print("leap year")
elif year % 4 == 0 and year % 100 != 0:
  print("leap year")
else:
  print("common year")
```
